# Q.2 IMDB’s Top rated 100 movies

```python
imdb= requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
imdb
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
imdb= requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
imdb
```

```python
soup= BeautifulSoup(imdb.content)
soup
```

```python
#Top 100 Rated Movies
name= soup.find_all("h3", class_="lister-item-header")
name
```

```python
len(name)
```

```python
series_name=[] 

for i in name:
    series_name.append(i.text.replace("\n",""))
    
series_name    
```

```python
series_name= [] 

for i in name:
    for j in i.find_all("a"):
        series_name.append(j.text.replace("\n",""))
    
series_name
```

```python
rating= soup.find_all("div", class_="inline-block ratings-imdb-rating")
rating
```

```python
len(rating)
```

```python
Rating=[] 

for i in rating:
    Rating.append(i.text.replace("\n",""))
    
Rating
```

```python
year= soup.find_all("span", class_="lister-item-year text-muted unbold")
year
```

```python
Year=[] 

for i in year:
    Year.append(i.text)
    
Year
```

```python
import pandas as pd
```

```python
movies=pd.DataFrame({})
movies['series_name']=series_name
movies['Rating']=Rating
movies['Year']=Year

movies
```
